```python
import json
import re
import pandas as pd
```

```python
def deEmojify(inputString):
    return inputString.encode('ascii', 'ignore').decode('ascii')
```

```python
tweet_id = []
tweet_text = []
what_class = []
with open("tweets.json",'r',encoding = 'utf-8') as wr, open("tweets.txt", 'w',encoding = 'utf-8') as ww:
    data = json.load(wr)
    for data in data:
        tweet_id.append(data['tweet_id'])
        tweet_text.append(deEmojify(data['text']).replace("@", "").replace("#", ""))
        what_class.append('1')
df = pd.DataFrame(list(zip(tweet_id, tweet_text,what_class)), columns =['tweet_id', 'tweet','class'])
df.to_csv('tweets.csv',index=False)
```
